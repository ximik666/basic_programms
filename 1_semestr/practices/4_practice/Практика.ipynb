2. Напишите функцию change(lst), которая принимает список и меняет местами его первый и последний элемент и выводит полученный результат на экран. В исходном списке минимум 2 элемента.
3. Дан список некоторых целых чисел, который вводится с клавиатуры  1 2 3 4 5 20, найдите первое значение 20 в нем и, если оно присутствует, замените его на 200. Выведите список на экран. 
4. Напишите фунуцию sqr(lst), которая принимает список и возвращает список с квадратами чисел.
5. Дан список чисел, введенный с клавиатуры 1 2 3 4 5 6, необходимо удалить все вхождения числа 20 из него и вывести на экран. 
6. Вывести какие числа и по сколько раз встречаются в списке случайных чисел
7. Вывести элементы списка, которые больше своих соседей
8. Поменять местами самый большой и самый маленький элементы списка
