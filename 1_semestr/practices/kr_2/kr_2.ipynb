1. Перепишите файл, добавляя к каждой строке в начало номер строки по порядку. Задачу решите без создания временного файла

2. Напишите функцию check_file(file), которая проверяет существования файла. Если файл существует - функция возвращает число символов в файле (без символов переноса строк). Если же файл не существует - функция создает файл с именем file и вовзвращает 0.

3. Напишите функцию read_last(file, lines), которая будет открывать определенный файл file и выводить на печать построчно строки, начиная с середины файла в количестве lines (на всякий случай проверим, что задано положительное целое число). 
Протестируем функцию на файле «article.txt» со следующим содержимым:

*Вечерело
Жужжали мухи
Светил фонарик
Кипела вода в чайнике
Венера зажглась на небе
Деревья шумели
Тучи разошлись
Листва зеленела*
